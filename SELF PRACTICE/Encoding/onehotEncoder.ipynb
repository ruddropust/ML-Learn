### One hot Encoding with Top Categories

```python
count = df['brand'].value_counts()
print(count)
```

```python
df['brand'].nunique()
threshold = 100
```

```python
replace_values = count[count <= threshold].index
```

```python
pd.get_dummies(df['brand'].replace(replace_values,"uncommon"))
```
