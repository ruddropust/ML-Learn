```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.datasets import make_classification
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Step 1: Generate synthetic classification data with 2 features (data1, data2)
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, random_state=42)

# Step 2: Add polynomial features (degree=3) for the two features (data1, data2)
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)

# Step 3: Create a DataFrame with columns 'data1', 'data2', and 'output'
df = pd.DataFrame(X_poly[:, 1:3], columns=['data1', 'data2'])  # Taking just the first two polynomial features
df['output'] = y

# Display the first few rows of the DataFrame
df.head()
```

```python
sns.scatterplot(x='data1', y='data2', hue='output', data=df)
plt.legend(title='output', loc='upper left', labels=['0', '1'])
plt.show()
```

```python
x = df[['data1', 'data2']]  
y = df['output']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
print('Train accuracy:', model.score(x_test, y_test)*100)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=model, legend=2)
```
