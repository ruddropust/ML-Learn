```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier,plot_tree,DecisionTreeRegressor
```

```python
dataset = pd.read_csv(r'C:\Users\shark\Desktop\ML Learn\WSCubeTech Part 2\Dataset\Salary_Data.csv')
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isnull().sum()
```

```python
sns.pairplot(data=dataset)
```

show linearity in this plot

```python
x = dataset.iloc[:,:-1]
y = dataset["Salary"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
dt_reg = DecisionTreeRegressor()
dt_reg.fit(x_train,y_train)
```

```python
dt_reg.score(x_train,y_train) *100
```

```python
dt_reg.score(x_test,y_test)*100
```

```python
plt.figure(figsize=(40,40))
plot_tree(dt_reg,filled=True)
plt.savefig('DecisionTreeregression.png')
plt.show()
```
