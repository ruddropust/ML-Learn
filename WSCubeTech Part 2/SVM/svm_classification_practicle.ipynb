```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv(r"C:\Users\shark\Desktop\ML Learn\WSCubeTech Part 2\Dataset\placement.csv")
dataset.head(4)
```

```python
dataset.isnull().sum()
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="cgpa",y="score",data=dataset,hue="placed")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["placed"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=14)
```

```python
# kernal = ['linear', 'poly', 'rbf', 'sigmoid', 'precomputed']
```

```python
from sklearn.svm import SVC

sv = SVC(kernel='linear')
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100
```

Check overfit

```python
sv.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sv)
plt.show()
```
