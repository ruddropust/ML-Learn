```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_decision_regions
from sklearn.svm import SVR
```

```python
# dataset = pd.read_csv(r"C:\Users\shark\Desktop\ML Learn\WSCubeTech Part 2\Dataset\Placement_cgp.csv")
dataset = pd.read_csv(r"C:\Users\shark\Desktop\ML Learn\WSCubeTech Part 2\Dataset\polynomial_salary_data.csv")

dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
sns.scatterplot(x="Level",y="Salary",data=dataset)
```

```python
x = dataset[["Level"]]
y = dataset[["Salary"]]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
#  'linear', 'poly', 'rbf', 'sigmoid', 'precomputed'
```

```python
svr_c = SVR(kernel='poly')
svr_c.fit(x_train,y_train)
```

```python
svr_c.score(x_test,y_test)*100
```

```python
svr_c.score(x_train,y_train)*100
```

```python
sns.scatterplot(x="Level",y="Salary",data=dataset)
plt.plot(x,svr_c.predict(x),color='red')
plt.show()
```
