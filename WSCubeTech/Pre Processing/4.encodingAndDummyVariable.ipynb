## Encoding

```python
import pandas as pd
dataset = pd.read_csv("loan.csv")
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
dataset["Gender"].fillna(dataset["Gender"].mode()[0],inplace=True)
dataset.isnull().sum()
```

```python
dataset.isnull().info()
```

```python
#fill categorical null value
nullValueList = ["Gender","Married"]

for i in nullValueList:
    dataset[i].fillna(dataset[i].mode()[0],inplace=True)
```

```python
dataset.isnull().sum()
```

```python
en_data = dataset[nullValueList]
en_data.head()
```

```python
pd.get_dummies(en_data)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop="first")
dt = ohe.fit_transform(en_data).toarray()
```

```python
list_column = ["Gender_Male","Married_Yes"]
pd.DataFrame(dt,columns=list_column)
```
