## Logistic Regression

Logistic Regression is a statistical method for analyzing datasets in which there are one or more independent variables that determine an outcome. The outcome is measured with a dichotomous variable (in which there are only two possible outcomes). The goal of logistic regression is to find the best fitting model to describe the relationship between the dichotomous characteristic of interest (dependent variable) and a set of independent (predictor or explanatory) variables.  

Logistic regression is used for binary classification problems and the output of the logistic regression model is a probability that the given input point belongs to a certain class. The logistic function (also called the sigmoid function) is used to map the predicted values to probabilities. The logistic function is defined as:

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

Where:
- **σ(x)** is the logistic function.
- **e** is the base of the natural logarithm.
- **x** is the input to the function.

The logistic regression model can be written as:

$$
P(Y=1|X) = \sigma(\beta_0 + \beta_1X_1 + \beta_2X_2 + \ldots + \beta_nX_n)
$$

Where:
- **P(Y=1|X)** is the probability that the dependent variable **Y** equals 1 given the independent variables **X**.
- **β₀, β₁, …, βₙ** are the coefficients of the model.
- **X₁, X₂, …, Xₙ** are the independent variables.

### When to Use Logistic Regression

Logistic regression is used in the following scenarios:

1. **Binary Classification**: When the dependent variable is binary (i.e., it has two possible outcomes such as Yes/No, True/False, Success/Failure).

2. **Predicting Probabilities**: When you need to predict the probability of an event occurring, rather than just classifying the event.

3. **Relationship Between Variables**: When you want to understand the relationship between one dependent binary variable and one or more independent variables.

4. **Medical Research**: Commonly used in medical research to determine the likelihood of a patient having a disease based on various diagnostic measures.

5. **Marketing**: Used to predict whether a customer will buy a product or not based on various features like age, income, and previous purchase history.

6. **Credit Scoring**: Used by financial institutions to predict the likelihood of a borrower defaulting on a loan.

7. **Social Sciences**: Used to model binary outcomes in social sciences, such as whether a person will vote in an election or not.

Logistic regression is a powerful tool for binary classification problems and is widely used across various fields for its simplicity and interpretability.

### Sigmoid Function

The sigmoid function, also known as the logistic function, is defined as:

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

In this Jupyter Notebook, the sigmoid function is implemented as follows:

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Example usage
x = np.array([0, 2, -2])
print(sigmoid(x))
```

The variable `x` is a numpy array with the value `array([0, 2, -2])`. The sigmoid function applied to `x` will output the following:

```python
print(sigmoid(x))
```

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
````

````python
dataset = pd.read_csv('Social_Network_Ads.csv')
dataset.head(5)
````

````python
dataset.drop(['EstimatedSalary'], axis=1, inplace=True)
dataset.head(5)
````

````python
plt.figure(figsize=(4, 3))
sns.scatterplot(x='Age', y='Purchased', data=dataset)
# sns.lineplot(x='Age', y='Purchased', data=dataset)
````

````python
x= dataset[["Age"]]
y = dataset[["Purchased"]]
````

````python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
````

````python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(x_train, y_train)
````

````python
lr.score(x_test, y_test)*100
````

````python
lr.predict([[57]])
````

````python
plt.figure(figsize=(4, 3))
sns.scatterplot(x='Age', y='Purchased', data=dataset)
sns.lineplot(x='Age', y=lr.predict(x), data=dataset,color='red')    
````
